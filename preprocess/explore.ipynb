# Explore the dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# pre-processed, but not scaled (from earlier)
df = pd.read_csv('./output/spotify_preprocessed.csv')
df.head()
```

```python
numeric_cols = [
    "danceability", "energy", "acousticness", "instrumentalness", "valence", "tempo",
    "speechiness", "track_popularity", "artist_popularity", "artist_followers",
    "track_duration_ms", "explicit", "album_total_tracks", "album_release_date"
]

sns.heatmap(df[numeric_cols].corr(), cmap="coolwarm", center=0)
```

```python
sns.pairplot(df[numeric_cols].sample(1000))
```

```python
plt.figure(figsize=(10, 4))
plt.xticks(rotation=90)
sns.countplot(data=df, x='genre_fixed')
plt.title('Genre Distribution')
```
